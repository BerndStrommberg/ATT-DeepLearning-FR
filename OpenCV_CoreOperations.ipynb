```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Grundlegende Operationen mit Bildern

Könnte für uns durchaus gut sein die zu kennen. Da es hier darum geht, wie man Pixel anspricht etc.

Ich benutze den Rest des Dokuments die Funktion um Bilder einfach anzuzeigen.

```python
def drawPicture(name, image):
    cv2.imshow(name, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
img = cv2.imread("Bilder/Homer.png")
```

Pixel werden einfach über Reihe und Zeile angesprochen

```python
px = img[100,100]
print(px)
```

Genau so lassen sich die Pixelwerte auch beeinflussen.
Hier einmal das Ursprüngliche Bild:

```python
drawPicture("HomerFirst", img)
```

```python
img[100,100] = [0,0,255]
```

```python
drawPicture("HomerFirst", img) # Kann man nicht erkennen
```

```python
print(img[100,100])
```

```python
img[:,100] = [255,0,0]
```

```python
drawPicture("HomerFirst", img)
```

Um ganz Reihen anzusprechen macht die obere Art Sinn. Aber Numpy eignet sich da deutlich besser.

```python
img.item(100,100,2) # Der Rote Wert(Index 2) des Pixels an Position 100,100 hat den Wert 0
# Da wurde ja die Blaue Linie gezeichnet
```

```python
img.itemset((100,100,2),100)
img.item(100,100,2)
#Jetzt hat der Pixel auch einen ROten Werte
```

## Image ROI (Region Object Indexing? Keine Ahnung)

Hier benutzt man wieder Numpy Indexing.

```python
donut = img [60:100, 100:150]
drawPicture("Ausschnitt", donut)
```

```python
img[0:40, 0:50] = donut
drawPicture("Bearbeitet", img)
```

## Splitting und Merging von Bild Channels
So kann man zum Beispiel die Farbkanäle trennen und in der Reihenfolge RGB wieder zusammenfügen.

```python


b,g,r = cv2.split(img)
img = cv2.merge((r,g,b))
```

```python
plt.imshow(img, interpolation = 'bicubic')
plt.show
```

Man kann zum Beispiel auch alle Pixel eine Farbe schwarz färben.



```python
img[:, : , 0] = 255
```

```python
plt.imshow(img, interpolation = 'bicubic')
plt.show
```
