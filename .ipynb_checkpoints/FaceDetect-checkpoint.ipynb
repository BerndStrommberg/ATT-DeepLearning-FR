```python
from imutils import face_utils
import numpy as np
import argparse
import imutils
import dlib
import cv2
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline  
```

## Detector und Predictor laden

```python
detector = dlib.get_frontal_face_detector()
```

```python
predictor = dlib.shape_predictor("data/shape_predictor_68_face_landmarks.dat")
```

## Öffnen und Analysieren des Videos
Hier wird mithilfe von OpenCV das Video geöffnet und mit imutils skaliert. Dann wird es zu einem Graustufenbild konvertiert und schließlich gedreht. Die Landmarks, die in jedem Frame des Videos erkannt werden, werden in eine List gespeichert 

Nase ist die 30

```python
frames = []
frameIterator = 0
shapes = np.zeros((68,2))
```

```python
cap = cv2.VideoCapture('Bilder/Moritz2.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        image = imutils.resize(frame, width=500)

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        rows, cols = gray.shape

        M = cv2.getRotationMatrix2D((cols/2, rows/2), 180, 1)
        dst = cv2.warpAffine(gray, M, (cols,rows))
        
        
        
        rects = detector(dst, 1)
        
  
        
        for (i, rect) in enumerate(rects):
            # determine the facial landmarks for the face region, then
            # convert the facial landmark (x, y)-coordinates to a NumPy
            # array
            shape = predictor(dst, rect)
            shape = face_utils.shape_to_np(shape)
            shapes = shape[30] - shape
            #shapes[:,0] = shape[:,0] -shape[30,0]
            frames.append(shapes)
                
           
            (x, y, w, h) = face_utils.rect_to_bb(rect)
            cv2.rectangle(dst, (x, y), (x + w, y + h), (0, 255, 0), 2)
            for (x, y) in shape:
                cv2.circle(dst, (x, y), 1, (0, 0, 255), -1)
                #cv2.putText(dst,str(iterator),(x,y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0),1)
                #iterator += 1
        cv2.imshow('frame', dst)
    
        frameIterator += 1

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
        
    
cap.release()
cv2.destroyAllWindows()



```
