## Zugriff auf die Kamera

VideoCapture() bekommt als argument entweder den Namen des Videos oder den Index des Gerätes, das angesprochen werden soll. (0, -1 in der Regel)

```python
cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # Operations on the frame
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Display resultingframe (siehe oben)
    cv2.imshow('frame', gray)
    if(cv2.waitKey(1) & 0xFF == ord('q')):
        break
        

# Am Ende alles wieder schliessen und vor allem das capturing entlassen
cap.release()
cv2.destroyAllWindows()
```

Video Speichern


```python
cap = cv2.VideoCapture(0)
# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc, 20.0, (640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret == True:
        frame = cv2.flip(frame,0)
        
        #write the flipped frame
        out.write(frame)
        
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    else:
        break
        
# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```

## Drawing in OpenCv

Die Argumente, die dabei übergeben werden:
* img : The image where you want to draw the shapes
    color : Color of the shape. for BGR, pass it as a tuple, eg: (255,0,0) for blue. For grayscale, just pass the scalar value.
* thickness : Thickness of the line or circle etc. If -1 is passed for closed figures like circles, it will fill the shape. default thickness = 1
* lineType : Type of line, whether 8-connected, anti-aliased line etc. By default, it is 8-connected. cv2.LINE_AA gives anti-aliased line which looks great for curves.

```python
# Erstellt ein Array, schwarzes Bild
img = np.zeros((512,512,3), np.uint8)

# Diagonale blaue Linie
# Bild, Start, Ende, Farbe (BGR), Dicke
cv2.line(img, (0,0),(511,511), (255, 0, 0), 5)

# Viereck, Kreis, Ellipse
cv2.rectangle(img,(384,0),(510,128),(0,255,0),3)
cv2.circle(img,(447,63), 63, (0,0,255), -1)
cv2.ellipse(img,(256,256),(100,50),0,0,180,255,-1)

#Polygon
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(img,[pts],True,(0,255,255))

```

### Text auf Bild


```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, 'Shiaaaat', (10,500), font, 4, (255,255,255),2,cv2.LINE_AA)

# Bei Matplotlib ist die line rot, da das mit RGB arbeitet
plt.imshow(img,interpolation="bicubic")
#plt.xticks([]), plt.yticks([]) --> Damit lassen sich die Achsen verbergen
plt.show()


cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
